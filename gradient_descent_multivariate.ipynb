```python
def gradient_descent_2(alpha, x, y, numIterations):
    m = x.shape[0] # number of samples
    theta = np.ones(7) # forms an array of ones (1) with 2 columns


    x_transpose = x.transpose()
    for iter in range(0, numIterations):
        hypothesis = np.dot(x, theta)   ##Hypothesis is represented by h(x) = theta(zero) + theta(one)* X
        loss = hypothesis - y            ## Loss = Predicted value( hypothesis) - the actual value(y)
        J = np.sum(loss ** 2) / (2 * m)  # cost = 1/2*m(summation(loss)^2))
        print "iter %s | J: %.3f" % (iter, J)      
        gradient = np.dot(x_transpose, loss) / m ## gradient descent = 1/m((loss))* X        
        theta = theta - alpha * gradient  # update the theta
    return theta
```
